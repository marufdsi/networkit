```python
t1 = 48
t2 = 36
cascade_power_data_full = pd.read_csv('build/CCPE_Results/LM/PowerLog/PowerLog_Cascade.csv')
for i,r in cascade_power_data_full.iterrows():
    if r["Graph Name"] == "loc-brightkite_edges":
        cascade_power_data_full.at[i, "Graph Name"] = "loc-Brightkite"
    if r["Graph Name"] == "loc-gowalla_edges":
        cascade_power_data_full.at[i, "Graph Name"] = "loc-Gowalla"
    if r["Graph Name"] == "uk":
        cascade_power_data_full.at[i, "Graph Name"] = "uk-2002"
        
skylake_power_data_full = pd.read_csv('build/CCPE_Results/LM/PowerLog/PowerLog_Skylake.csv')
for i,r in skylake_power_data_full.iterrows():
    if r["Graph Name"] == "loc-brightkite_edges":
        skylake_power_data_full.at[i, "Graph Name"] = "loc-Brightkite"
    if r["Graph Name"] == "loc-gowalla_edges":
        skylake_power_data_full.at[i, "Graph Name"] = "loc-Gowalla"
    if r["Graph Name"] == "uk":
        skylake_power_data_full.at[i, "Graph Name"] = "uk-2002"
        
cascade_power_data_full = cascade_power_data_full[cascade_power_data_full['Iteration'] == 1].reset_index()
skylake_power_data_full = skylake_power_data_full[skylake_power_data_full['Iteration'] == 1].reset_index()
cascade_power_list = []
cascade_mplm_ovpl_power_list = []
skylake_power_list = []
skylake_mplm_ovpl_power_list = []
for graph in graph_list:
    cascade_power = cascade_power_data_full[(cascade_power_data_full["Power"] > 0) & (cascade_power_data_full["Threads"] == t1) & (cascade_power_data_full["Graph Name"] == graph)]
    cascade_mplmPower = cascade_power[cascade_power["Version"] == "MPLM"]
    cascade_onplPower = cascade_power[cascade_power["Version"] == "ONPL"]
    cascade_ovplPower = cascade_power[cascade_power["Version"] == "OVPL"]
    cascade_mplmEnergy = 0
    cascade_onplEnergy = 0
    cascade_ovplEnergy = 0
    cascade_pwr_group = cascade_mplmPower.groupby("Event")
    for event,pwr_data in cascade_pwr_group:
        if event in ["package-0", "package-1", "dram"]:
            cascade_mplmEnergy += pwr_data.mean()["Power"]
    cascade_pwr_group = cascade_onplPower.groupby("Event")
    for event,pwr_data in cascade_pwr_group:
        if event in ["package-0", "package-1", "dram"]:
            cascade_onplEnergy += pwr_data.mean()["Power"]
    for event,pwr_data in cascade_ovplPower.groupby("Event"):
        if event in ["package-0", "package-1", "dram"]:
            cascade_ovplEnergy += pwr_data.mean()["Power"]
    if cascade_onplEnergy> 0 and cascade_mplmEnergy > 0:
        cascade_power_list.append((cascade_mplmEnergy/cascade_onplEnergy))
    else:
        cascade_power_list.append(0)
    if cascade_ovplEnergy> 0 and cascade_mplmEnergy > 0:
        cascade_mplm_ovpl_power_list.append((cascade_mplmEnergy/cascade_ovplEnergy))
    else:
        cascade_mplm_ovpl_power_list.append(0)
    
    skylake_power = skylake_power_data_full[(skylake_power_data_full["Power"] > 0) & (skylake_power_data_full["Threads"] == t2) & (skylake_power_data_full["Graph Name"] == graph)]
    skylake_mplmPower = skylake_power[skylake_power["Version"] == "MPLM"]
    skylake_onplPower = skylake_power[skylake_power["Version"] == "ONPL"]
    skylake_ovplPower = cascade_power[cascade_power["Version"] == "OVPL"]
    skylake_mplmEnergy = 0
    skylake_onplEnergy = 0
    skylake_ovplEnergy = 0
    skylake_pwr_group = skylake_mplmPower.groupby("Event")
    for event,pwr_data in skylake_pwr_group:
        if event in ["package-0", "package-1", "dram"]:
            skylake_mplmEnergy += pwr_data.mean()["Power"]
    skylake_pwr_group = skylake_onplPower.groupby("Event")
    for event,pwr_data in skylake_pwr_group:
        if event in ["package-0", "package-1", "dram"]:
            skylake_onplEnergy += pwr_data.mean()["Power"]
    for event,pwr_data in skylake_ovplPower.groupby("Event"):
        if event in ["package-0", "package-1", "dram"]:
            skylake_ovplEnergy += pwr_data.mean()["Power"]
    if skylake_mplmEnergy > 0 and skylake_onplEnergy >0:
        skylake_power_list.append(skylake_mplmEnergy/skylake_onplEnergy)
    else:
        skylake_power_list.append(0)
    if skylake_mplmEnergy > 0 and skylake_ovplEnergy >0:
        skylake_mplm_ovpl_power_list.append(skylake_mplmEnergy/skylake_ovplEnergy)
    else:
        skylake_mplm_ovpl_power_list.append(0)
    
    
    
lvData  = pd.DataFrame()
lvData["Graph"] = graph_list
lvData["Cascade Lake"] = cascade_power_list
lvData["SkyLake"] = skylake_power_list

X = graph_list
X_axis = np.arange(len(X))
fig, ax = plt.subplots(figsize=(20,8))
plt.bar(X_axis - 0.2, cascade_power_list, 0.4, label = 'Cascade Lake', hatch='///', rasterized=True)
plt.bar(X_axis + 0.2, skylake_power_list, 0.4, label = 'Skylake', hatch='---', rasterized=True)

plt.xticks(X_axis, X, rotation=90)
plt.xlabel("")
plt.ylabel("Energy Ratio(Scalar/Vectorized)", fontsize=20)
plt.grid(axis="y")
plt.axhline(1, linestyle='--', linewidth=2.0, color="red")
plt.legend(fontsize=20)
plt.tick_params(axis='x', colors='#800000', labelsize='20')
plt.savefig("../TestPLMFromNetworkIt/paper/CCPE21/figures/louvain/power/mplm_vs_onpl_power_cascade_" + str(t1) + "_skl_" + str(t2) + "_mean.pdf", bbox_inches = 'tight') 


df  = pd.DataFrame()
df["Graph"] = graph_list
df["Cascade Lake"] = cascade_mplm_ovpl_power_list
df["SkyLake"] = skylake_mplm_ovpl_power_list

X = graph_list
X_axis = np.arange(len(X))
fig, ax = plt.subplots(figsize=(20,8))
plt.bar(X_axis - 0.2, cascade_mplm_ovpl_power_list, 0.4, label = 'Cascade Lake', hatch='///', rasterized=True)
plt.bar(X_axis + 0.2, skylake_mplm_ovpl_power_list, 0.4, label = 'Skylake', hatch='---', rasterized=True)

plt.xticks(X_axis, X, rotation=90)
plt.xlabel("")
plt.ylabel("Energy Ration(Scalar/Vectorized)", fontsize=20)
plt.grid(axis="y")
plt.axhline(1, linestyle='--', linewidth=2.0, color="red")
plt.legend(fontsize=20)
plt.tick_params(axis='x', colors='#800000', labelsize='20')

plt.savefig("../TestPLMFromNetworkIt/paper/CCPE21/figures/louvain/power/mplm_vs_ovpl_power_cascade_" + str(t1) + "_skl_" + str(t2) + "_mean.pdf", bbox_inches = 'tight') 
```

```python
t1 = 48
t2 = 36
cascade_power_data_full = pd.read_csv('../../EUPAR_Results/PowerLog/PowerLog_Cascade.csv')
for i,r in cascade_power_data_full.iterrows():
    if r["Graph Name"] == "loc-brightkite_edges":
        cascade_power_data_full.at[i, "Graph Name"] = "loc-Brightkite"
    if r["Graph Name"] == "uk":
        cascade_power_data_full.at[i, "Graph Name"] = "uk-2002"
skylake_power_data_full = pd.read_csv('../../EUPAR_Results/PowerLog/PowerLog_Skylake.csv')
for i,r in skylake_power_data_full.iterrows():
    if r["Graph Name"] == "loc-brightkite_edges":
        skylake_power_data_full.at[i, "Graph Name"] = "loc-Brightkite"
    if r["Graph Name"] == "uk":
        skylake_power_data_full.at[i, "Graph Name"] = "uk-2002"
cascade_power_data_full = cascade_power_data_full[cascade_power_data_full['Iteration'] == 1].reset_index()
skylake_power_data_full = skylake_power_data_full[skylake_power_data_full['Iteration'] == 1].reset_index()
cascade_power_list = []
cascade_mplm_ovpl_power_list = []
skylake_power_list = []
skylake_mplm_ovpl_power_list = []
for graph in graph_list:
    cascade_power = cascade_power_data_full[(cascade_power_data_full["Power"] > 0) & (cascade_power_data_full["Threads"] == t1) & (cascade_power_data_full["Graph Name"] == graph)]
    cascade_mplmPower = cascade_power[cascade_power["Version"] == "Modified Legacy"]
    cascade_onplPower = cascade_power[cascade_power["Version"] == "Vectorized Row PLM"]
    cascade_ovplPower = cascade_power[cascade_power["Version"] == "Vectorized Block PLM"]
    cascade_mplmEnergy = 0
    cascade_onplEnergy = 0
    cascade_ovplEnergy = 0
    cascade_pwr_group = cascade_mplmPower.groupby("Event")
    for event,pwr_data in cascade_pwr_group:
        if event in ["package-0", "package-1", "dram"]:
            cascade_mplmEnergy += pwr_data.mean()["Power"]
    cascade_pwr_group = cascade_onplPower.groupby("Event")
    for event,pwr_data in cascade_pwr_group:
        if event in ["package-0", "package-1", "dram"]:
            cascade_onplEnergy += pwr_data.mean()["Power"]
    for event,pwr_data in cascade_ovplPower.groupby("Event"):
        if event in ["package-0", "package-1", "dram"]:
            cascade_ovplEnergy += pwr_data.mean()["Power"]
    if cascade_onplEnergy> 0 and cascade_mplmEnergy > 0:
        cascade_power_list.append((cascade_mplmEnergy/cascade_onplEnergy))
    else:
        cascade_power_list.append(0)
    if cascade_ovplEnergy> 0 and cascade_mplmEnergy > 0:
        cascade_mplm_ovpl_power_list.append((cascade_mplmEnergy/cascade_ovplEnergy))
    else:
        cascade_mplm_ovpl_power_list.append(0)
    
    skylake_power = skylake_power_data_full[(skylake_power_data_full["Power"] > 0) & (skylake_power_data_full["Threads"] == t2) & (skylake_power_data_full["Graph Name"] == graph)]
    skylake_mplmPower = skylake_power[skylake_power["Version"] == "Modified Legacy"]
    skylake_onplPower = skylake_power[skylake_power["Version"] == "Vectorized Row PLM"]
    skylake_ovplPower = cascade_power[cascade_power["Version"] == "Vectorized Block PLM"]
    skylake_mplmEnergy = 0
    skylake_onplEnergy = 0
    skylake_ovplEnergy = 0
    skylake_pwr_group = skylake_mplmPower.groupby("Event")
    for event,pwr_data in skylake_pwr_group:
        if event in ["package-0", "package-1", "dram"]:
            skylake_mplmEnergy += pwr_data.mean()["Power"]
    skylake_pwr_group = skylake_onplPower.groupby("Event")
    for event,pwr_data in skylake_pwr_group:
        if event in ["package-0", "package-1", "dram"]:
            skylake_onplEnergy += pwr_data.mean()["Power"]
    for event,pwr_data in skylake_ovplPower.groupby("Event"):
        if event in ["package-0", "package-1", "dram"]:
            skylake_ovplEnergy += pwr_data.mean()["Power"]
    if skylake_mplmEnergy > 0 and skylake_onplEnergy >0:
        skylake_power_list.append(skylake_mplmEnergy/skylake_onplEnergy)
    else:
        skylake_power_list.append(0)
    if skylake_mplmEnergy > 0 and skylake_ovplEnergy >0:
        skylake_mplm_ovpl_power_list.append(skylake_mplmEnergy/skylake_ovplEnergy)
    else:
        skylake_mplm_ovpl_power_list.append(0)
    
    
    
lvData  = pd.DataFrame()
lvData["Graph"] = graph_list
lvData["Cascade Lake"] = cascade_power_list
lvData["SkyLake"] = skylake_power_list

X = graph_list
X_axis = np.arange(len(X))
fig, ax = plt.subplots(figsize=(20,8))
plt.bar(X_axis - 0.2, cascade_power_list, 0.4, label = 'Cascade Lake', hatch='///', rasterized=True)
plt.bar(X_axis + 0.2, skylake_power_list, 0.4, label = 'Skylake', hatch='---', rasterized=True)

plt.xticks(X_axis, X, rotation=90)
plt.xlabel("")
plt.ylabel("Energy Ration(Scalar/Vectorized)", fontsize=20)
plt.grid(axis="y")
plt.axhline(1, linestyle='--', linewidth=2.0, color="red")
plt.legend(fontsize=20)
plt.tick_params(axis='x', colors='#800000', labelsize='20')
plt.savefig("figures/louvain/mplm_vs_onpl_power_cascade_" + str(t1) + "_skl_" + str(t2) + "_mean.pdf", bbox_inches = 'tight') 


df  = pd.DataFrame()
df["Graph"] = graph_list
df["Cascade Lake"] = cascade_mplm_ovpl_power_list
df["SkyLake"] = skylake_mplm_ovpl_power_list

X = graph_list
X_axis = np.arange(len(X))
fig, ax = plt.subplots(figsize=(20,8))
plt.bar(X_axis - 0.2, cascade_mplm_ovpl_power_list, 0.4, label = 'Cascade Lake', hatch='///', rasterized=True)
plt.bar(X_axis + 0.2, skylake_mplm_ovpl_power_list, 0.4, label = 'Skylake', hatch='---', rasterized=True)

plt.xticks(X_axis, X, rotation=90)
plt.xlabel("")
plt.ylabel("Energy Ration(Scalar/Vectorized)", fontsize=20)
plt.grid(axis="y")
plt.axhline(1, linestyle='--', linewidth=2.0, color="red")
plt.legend(fontsize=20)
plt.tick_params(axis='x', colors='#800000', labelsize='20')

plt.savefig("figures/louvain/mplm_vs_ovpl_power_cascade_" + str(t1) + "_skl_" + str(t2) + "_mean.pdf", bbox_inches = 'tight') 
```
